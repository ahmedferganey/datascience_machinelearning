```python
import pandas as pd
import csv
```

```python
data_dicts = [
    {'Name': 'Alice'    , 'Age': 30     , 'City': 'New York'    },
    {'Name': 'Bob'      , 'Age': 25     , 'City': 'Los Angeles' },
    {'Name': 'Charlie'  , 'Age': 35     , 'City': 'Chicago'     }
]
output_filename = 'output_dict.csv'

with open(output_filename, 'w', newline='') as file:
    fieldnames = ['Name', 'Age', 'City']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()  # Write the header
    writer.writerows(data_dicts)

# Step 2: Detect the delimiter of the CSV file
with open(output_filename, 'r') as file:
    sample = file.read(1024)  # Read a sample of the file
    sniffer = csv.Sniffer()   #The csv.Sniffer() is a class in Python's csv module that helps you deduce the format of a CSV file. Specifically, it can detect the delimiter and whether the first row in the file contains headers.
    detected_delimiter = sniffer.sniff(sample).delimiter

print(f"The detected delimiter in '{output_filename}' is: '{detected_delimiter}'")
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid")
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data_visulization = pd.read_csv('/media/ahmed-ferganey/AI4/01-Learning_AI/MyGitHub/datascience_machinelearning/Airline_Delay.csv')
```

```python
data_visulization['carrier'].nunique()
```

```python
data_visulization
```

```python
def Pie(feature, Limit=10):
    global data_visulization
    fig, ax = plt.subplots(figsize=(10, 6))
    # Get the value counts and slice it based on Limit
    value_counts = data_visulization[feature].value_counts()[:Limit]
    plt.pie(value_counts, 
            labels=value_counts.index,
            autopct='%1.2f%%',
            labeldistance=1.1,
            explode=[0.05 for _ in range(len(value_counts))])
    plt.show()

# Call the function with 'carrier'
Pie('carrier')
```

```python

def SelectedFeaturePie(feature, Limit=10):
    Feature = 'year'
    Values = ['2019', '2020', '2021']
    global data_visulization
    
    fig, ax = plt.subplots(figsize=(15, 4))
    
    for i, value in enumerate(Values):
        plt.subplot(1, 3, i+1)
        plt.title(value)
        
        Data = data_visulization[data_visulization[Feature] == value]
        
        if Data.empty:
            print(f"No data for {value}")
            continue

        value_counts = Data[feature].value_counts()[:Limit]
        if value_counts.empty:
            print(f"No values to plot for {value}")
            continue

        plt.pie(
            value_counts,
            labels=value_counts.index,
            autopct='%1.2f%%',
            labeldistance=1.1,
            explode=[0.05 for _ in range(len(value_counts))]
        )
    
    plt.show()

# Call the function
SelectedFeaturePie('carrier', Limit=17)
```

```python

def SelectedFeaturePie2(feature,Limit=10) : 
    Feature = 'year'
    Values = ['2019', '2020', '2021']
    global data_visulization
    fig, ax = plt.subplots(figsize=(15,4))
    
    plt.subplot(1,3,1)
    plt.title(Values[0])
    Data = data_visulization[data_visulization[Feature]==Values[0]]
    plt.pie(Data[feature].value_counts()[:Limit],labels=list(
        Data[feature].value_counts()[:Limit].index),
        autopct ='%1.2f%%' , labeldistance = 1.1,
            explode = [0.05 for i in range(len(Data[feature].value_counts()[:Limit]))] )
  
    plt.subplot(1,3,2)
    plt.title(Values[1])
    Data = data_visulization[data_visulization[Feature]==Values[1]]
    plt.pie(Data[feature].value_counts()[:Limit],labels=list(Data[feature].value_counts()[:Limit].index),
        autopct ='%1.2f%%' , labeldistance = 1.1,explode = [0.05 for i in range(len(Data[feature].value_counts()[:Limit]))] )

    plt.subplot(1,3,3)
    plt.title(Values[2])
    Data = data_visulization[data_visulization[Feature]==Values[2]]
    plt.pie(Data[feature].value_counts()[:Limit],labels=list(Data[feature].value_counts()[:Limit].index),
        autopct ='%1.2f%%' , labeldistance = 1.1,explode = [0.05 for i in range(len(Data[feature].value_counts()[:Limit]))] )
    plt.show()   

SelectedFeaturePie2('year', Limit=17)
  
```

```python
def JPlot(feature_1, feature_2):
    global data_visulization
    # Create a joint plot with KDE
    sns.jointplot(x=feature_1, y=feature_2, data=data_visulization, kind='kde')
    plt.show()

# Call the function
JPlot('arr_del15', 'security_ct')
```
